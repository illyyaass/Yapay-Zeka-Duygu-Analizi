Stemming

Stemming is the procisse of reducing a word its root word

exampel: actor, actress, acting= act

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]', ' ', content)

    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()

    stemmed_content = [port_stem.stem(word) for word in stemmed_content if word not in stopwords.words('english')]

    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
twitter_data['stemmed_content'] = twitter_data['text'].apply(stemming)
```

```python
import gradio as gr

# 1. Fonction d'analyse (Nafs l-logic)
def predict_sentiment(text):
    clean_text = stemming(text)
    vectorized_text = vectorizer.transform([clean_text])
    prediction_prob = model.predict_proba(vectorized_text)[0]

    # Rddina n-natija b Torkiya
    return {
        "Olumsuz 😔 (Negative)": float(prediction_prob[0]),
        "Olumlu 😃 (Positive)": float(prediction_prob[1])
    }

# 2. Theme Ihtirafi (Soft & Cyan)
my_theme = gr.themes.Soft(
    primary_hue="cyan",
    secondary_hue="slate",
    neutral_hue="slate",
    font=[gr.themes.GoogleFont('Poppins'), 'ui-sans-serif', 'system-ui']
)

# 3. CSS (Styling)
custom_css = """
h1 {
    text-align: center;
    color: #0891b2;
    font-weight: 800;
    font-size: 2.5rem;
    margin-bottom: 10px;
}
.subtitle {
    text-align: center;
    color: #64748b;
    font-size: 1.1rem;
    margin-bottom: 20px;
}
"""

# 4. Bina' l-Interface b Torkiya
with gr.Blocks(theme=my_theme, css=custom_css, title="Yapay Zeka Duygu Analizi") as app:

    # Basliklar (Headers)
    gr.Markdown("# 🤖 Yapay Zeka Duygu Analizi")
    gr.Markdown("<div class='subtitle'>İngilizce tweetler için duygu analizi modeli (Sentiment Analysis)</div>")

    # Layout: Row & Columns
    with gr.Row():

        # Column 1: Giris (Input)
        with gr.Column(scale=1):
            input_text = gr.Textbox(
                label="Metin Giriniz (İngilizce)",
                placeholder="Örnek: I really enjoyed the service, it was fantastic! ...",
                lines=5,
                elem_id="input-box"
            )
            # Butonlar
            with gr.Row():
                submit_btn = gr.Button("🔍 Analiz Et", variant="primary", size="lg")
                clear_btn = gr.Button("🗑️ Temizle", variant="secondary")

        # Column 2: Sonuc (Output)
        with gr.Column(scale=1):
            output_label = gr.Label(
                label="Analiz Sonucu",
                num_top_classes=2
            )

    # Actions
    submit_btn.click(fn=predict_sentiment, inputs=input_text, outputs=output_label)
    clear_btn.click(lambda: (None, None), inputs=None, outputs=[input_text, output_label])

# 5. App'i baslat
app.launch()
```
